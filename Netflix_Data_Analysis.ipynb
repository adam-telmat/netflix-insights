```python
#19
#Sans print avec DataFrame propre: 
import pandas as pd

# Charger les données
df = pd.read_csv("C:/Users/claoy/Desktop/netflix_titles.csv")

# Si listed_in n'est pas encore transformée en liste, on le fait
if isinstance(df['listed_in'].iloc[0], str):
    df['listed_in'] = df['listed_in'].apply(lambda x: x.split(', ') if pd.notna(x) else [])

# Colonnes à analyser
colonnes = ['type', 'country', 'release_year', 'rating', 'listed_in']

# Construction du tableau résumé
résumé = {
    'Variable': [],
    'Extrait des valeurs uniques': [],
    'Nombre de valeurs uniques': []
}

for col in colonnes:
    if col == 'listed_in':
        # Aplatir toutes les listes
        valeurs_uniques = pd.Series([genre for sublist in df[col] if isinstance(sublist, list) for genre in sublist])
    else:
        valeurs_uniques = df[col].dropna().unique()
        valeurs_uniques = pd.Series(valeurs_uniques)
    
    résumé['Variable'].append(col)
    extrait = ', '.join(map(str, sorted(valeurs_uniques.unique()[:5]))) + ('...' if len(valeurs_uniques.unique()) > 5 else '')
    résumé['Extrait des valeurs uniques'].append(extrait)
    résumé['Nombre de valeurs uniques'].append(valeurs_uniques.nunique())

# Afficher le tout sous forme de DataFrame
résumé_df = pd.DataFrame(résumé)
résumé_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement du dataset
df = pd.read_csv("C:/Users/claoy/Desktop/netflix_titles.csv")

# a. Répartition du type d'œuvres
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type')
plt.title("Répartition des types d'œuvres")
plt.xlabel("Type")
plt.ylabel("Nombre")
plt.show()

# b. Répartition des œuvres par pays (Top 10)
plt.figure(figsize=(10, 6))
df['country'].dropna().str.split(', ').explode().value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 des pays avec le plus d'œuvres")
plt.xlabel("Pays")
plt.ylabel("Nombre d'œuvres")
plt.xticks(rotation=45)
plt.show()

# c. Répartition des œuvres par année d'ajout
df['year_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.year
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='year_added', order=sorted(df['year_added'].dropna().unique()))
plt.title("Répartition des œuvres par année d'ajout")
plt.xlabel("Année")
plt.ylabel("Nombre")
plt.xticks(rotation=45)
plt.show()

# d. Répartition des ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index)
plt.title("Répartition des ratings")
plt.xlabel("Nombre d'œuvres")
plt.ylabel("Rating")
plt.show()

# e. Répartition de la durée des films
film_durations = df[df['type'] == 'Movie']['duration'].dropna()
film_durations = film_durations[film_durations.str.contains('min')].str.replace(' min', '').astype(int)

plt.figure(figsize=(10, 5))
sns.histplot(film_durations, bins=30, kde=True, color='orange')
plt.title("Répartition des durées des films (en minutes)")
plt.xlabel("Durée")
plt.ylabel("Nombre de films")
plt.show()

# f. Répartition de la durée des séries
series_seasons = df[df['type'] == 'TV Show']['duration'].dropna()
series_seasons = series_seasons.str.replace(' Season', '').str.replace('s', '').astype(int)

plt.figure(figsize=(8, 5))
sns.countplot(x=series_seasons)
plt.title("Répartition des séries par nombre de saisons")
plt.xlabel("Nombre de saisons")
plt.ylabel("Nombre de séries")
plt.show()

# g. Répartition des genres d'œuvres
plt.figure(figsize=(10, 6))
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='barh', color='purple')
plt.title("Top 10 des genres les plus représentés")
plt.xlabel("Nombre")
plt.ylabel("Genre")
plt.gca().invert_yaxis()
plt.show()

# h. Top 5 des séries les plus longues
top_series = df[df['type'] == 'TV Show'].copy()
top_series['num_seasons'] = top_series['duration'].str.extract(r'(\d+)').astype(float)
top_series = top_series.sort_values(by='num_seasons', ascending=False).head(5)

plt.figure(figsize=(8, 4))
sns.barplot(data=top_series, x='num_seasons', y='title')
plt.title("Top 5 des séries les plus longues (saisons)")
plt.xlabel("Nombre de saisons")
plt.ylabel("Série")
plt.show()

# i. Top 5 des films les plus longs
top_movies = df[df['type'] == 'Movie'].copy()
top_movies['duration_min'] = top_movies['duration'].str.extract(r'(\d+)').astype(float)
top_movies = top_movies.sort_values(by='duration_min', ascending=False).head(5)

plt.figure(figsize=(8, 4))
sns.barplot(data=top_movies, x='duration_min', y='title')
plt.title("Top 5 des films les plus longs")
plt.xlabel("Durée (minutes)")
plt.ylabel("Film")
plt.show()

# j. Répartition des réalisateurs des œuvres françaises
fr_works = df[df['country'].str.contains("France", na=False)]
plt.figure(figsize=(10, 6))
fr_works['director'].dropna().value_counts().head(10).plot(kind='bar', color='darkred')
plt.title("Top 10 des réalisateurs d'œuvres françaises")
plt.xlabel("Réalisateur")
plt.ylabel("Nombre d'œuvres")
plt.xticks(rotation=45)
plt.show()

# k. Répartition des œuvres par date d'ajout au catalogue
date_counts = pd.to_datetime(df['date_added'], errors='coerce').dt.date.value_counts().sort_index()

plt.figure(figsize=(14, 5))
date_counts.plot()
plt.title("Répartition des ajouts au catalogue au fil du temps")
plt.xlabel("Date d'ajout")
plt.ylabel("Nombre d'ajouts")
plt.tight_layout()
plt.show()

# l. Durée moyenne des films selon le genre
films = df[df['type'] == 'Movie'].copy()
films['duration_min'] = films['duration'].str.extract(r'(\d+)').astype(float)
films['genre'] = films['listed_in'].str.split(', ').str[0]

plt.figure(figsize=(12, 5))
films.groupby('genre')['duration_min'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='teal')
plt.title("Durée moyenne des films selon le genre")
plt.ylabel("Durée moyenne (min)")
plt.xlabel("Genre")
plt.xticks(rotation=45)
plt.show()
```

### Analyse des graphiques de visualisation du dataset Netflix

**a. Type d'œuvres :**
Le dataset contient principalement des films, mais les séries représentent aussi une part significative. Cela montre que Netflix mise sur les deux formats, avec toutefois une prédominance des films.

**b. Pays d'origine :**
Les œuvres proviennent majoritairement des États-Unis, suivis de l'Inde, du Royaume-Uni et du Canada. La France arrive en cinquième position. Cela démontre une forte dominance des pays anglophones, avec une certaine diversité internationale.

**c. Année d'ajout au catalogue :**
On observe une croissance importante des ajouts à partir de 2016, avec un pic en 2020. Cela correspond à la stratégie de croissance de Netflix, renforcée par la crise du COVID-19.

**d. Ratings :**
Les œuvres sont principalement destinées à un public adulte (TV-MA) ou adolescent (TV-14). Les contenus pour enfants ou familles sont moins fréquents.

**e. Durée des films :**
La plupart des films durent entre 80 et 120 minutes, ce qui correspond aux standards classiques du cinéma. Peu de films dépassent les 2 heures.

**f. Durée des séries (saisons) :**
La majorité des séries ont une ou deux saisons. Il y a peu de séries très longues, ce qui suggère une rotation rapide du contenu ou des formats courts.

**g. Genres :**
Les genres les plus représentés sont "Dramas", "Comedies" et "International Movies". Cela montre un intérêt pour les contenus narratifs et diversifiés culturellement.

**h. Top 5 des séries les plus longues :**
Certaines séries atteignent jusqu'à 15 saisons. Ces cas restent rares mais montrent que quelques titres phares restent longtemps dans le catalogue.

**i. Top 5 des films les plus longs :**
Des films de plus de 200 minutes existent, mais ils sont exceptionnels. Cela confirme que Netflix préfère le format moyen pour le grand public.

**j. Réalisateurs français :**
Quelques réalisateurs se démarquent avec plusieurs œuvres disponibles. Cela peut révéler des partenariats entre Netflix et certains auteurs locaux.

**k. Date d'ajout :**
Les ajouts sont réguliers et en augmentation jusqu'à 2020. Ensuite, on observe un ralentissement. Cela peut être lié à des changements stratégiques ou à l'effet post-pandémie.

**l. Durée moyenne des films par genre :**
Certains genres comme le drame ou le documentaire présentent des durées moyennes plus longues. Cela reflète des besoins narratifs plus complexes.

**Conclusion générale :**
Netflix dispose d'un catalogue diversifié, majoritairement composé de films et fortement orienté vers les pays anglophones. Les ajouts ont été très dynamiques entre 2016 et 2020, avec une stratégie centrée sur des formats accessibles et populaires.
