```python
#19
#Sans print avec DataFrame propre: 
import pandas as pd

# Charger les donn√©es
df = pd.read_csv("C:/Users/claoy/Desktop/netflix_titles.csv")

# Si listed_in n'est pas encore transform√©e en liste, on le fait
if isinstance(df['listed_in'].iloc[0], str):
    df['listed_in'] = df['listed_in'].apply(lambda x: x.split(', ') if pd.notna(x) else [])

# Colonnes √† analyser
colonnes = ['type', 'country', 'release_year', 'rating', 'listed_in']

# Construction du tableau r√©sum√©
r√©sum√© = {
    'Variable': [],
    'Extrait des valeurs uniques': [],
    'Nombre de valeurs uniques': []
}

for col in colonnes:
    if col == 'listed_in':
        # Aplatir toutes les listes
        valeurs_uniques = pd.Series([genre for sublist in df[col] if isinstance(sublist, list) for genre in sublist])
    else:
        valeurs_uniques = df[col].dropna().unique()
        valeurs_uniques = pd.Series(valeurs_uniques)
    
    r√©sum√©['Variable'].append(col)
    extrait = ', '.join(map(str, sorted(valeurs_uniques.unique()[:5]))) + ('...' if len(valeurs_uniques.unique()) > 5 else '')
    r√©sum√©['Extrait des valeurs uniques'].append(extrait)
    r√©sum√©['Nombre de valeurs uniques'].append(valeurs_uniques.nunique())

# Afficher le tout sous forme de DataFrame
r√©sum√©_df = pd.DataFrame(r√©sum√©)
r√©sum√©_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement du dataset
df = pd.read_csv("C:/Users/claoy/Desktop/netflix_titles.csv")

# a. R√©partition du type d'≈ìuvres
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type')
plt.title("R√©partition des types d'≈ìuvres")
plt.xlabel("Type")
plt.ylabel("Nombre")
plt.show()

# b. R√©partition des ≈ìuvres par pays (Top 10)
plt.figure(figsize=(10, 6))
df['country'].dropna().str.split(', ').explode().value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 des pays avec le plus d'≈ìuvres")
plt.xlabel("Pays")
plt.ylabel("Nombre d'≈ìuvres")
plt.xticks(rotation=45)
plt.show()

# c. R√©partition des ≈ìuvres par ann√©e d'ajout
df['year_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.year
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='year_added', order=sorted(df['year_added'].dropna().unique()))
plt.title("R√©partition des ≈ìuvres par ann√©e d'ajout")
plt.xlabel("Ann√©e")
plt.ylabel("Nombre")
plt.xticks(rotation=45)
plt.show()

# d. R√©partition des ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index)
plt.title("R√©partition des ratings")
plt.xlabel("Nombre d'≈ìuvres")
plt.ylabel("Rating")
plt.show()

# e. R√©partition de la dur√©e des films
film_durations = df[df['type'] == 'Movie']['duration'].dropna()
film_durations = film_durations[film_durations.str.contains('min')].str.replace(' min', '').astype(int)

plt.figure(figsize=(10, 5))
sns.histplot(film_durations, bins=30, kde=True, color='orange')
plt.title("R√©partition des dur√©es des films (en minutes)")
plt.xlabel("Dur√©e")
plt.ylabel("Nombre de films")
plt.show()

# f. R√©partition de la dur√©e des s√©ries
series_seasons = df[df['type'] == 'TV Show']['duration'].dropna()
series_seasons = series_seasons.str.replace(' Season', '').str.replace('s', '').astype(int)

plt.figure(figsize=(8, 5))
sns.countplot(x=series_seasons)
plt.title("R√©partition des s√©ries par nombre de saisons")
plt.xlabel("Nombre de saisons")
plt.ylabel("Nombre de s√©ries")
plt.show()

# g. R√©partition des genres d'≈ìuvres
plt.figure(figsize=(10, 6))
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='barh', color='purple')
plt.title("Top 10 des genres les plus repr√©sent√©s")
plt.xlabel("Nombre")
plt.ylabel("Genre")
plt.gca().invert_yaxis()
plt.show()

# h. Top 5 des s√©ries les plus longues
top_series = df[df['type'] == 'TV Show'].copy()
top_series['num_seasons'] = top_series['duration'].str.extract(r'(\d+)').astype(float)
top_series = top_series.sort_values(by='num_seasons', ascending=False).head(5)

plt.figure(figsize=(8, 4))
sns.barplot(data=top_series, x='num_seasons', y='title')
plt.title("Top 5 des s√©ries les plus longues (saisons)")
plt.xlabel("Nombre de saisons")
plt.ylabel("S√©rie")
plt.show()

# i. Top 5 des films les plus longs
top_movies = df[df['type'] == 'Movie'].copy()
top_movies['duration_min'] = top_movies['duration'].str.extract(r'(\d+)').astype(float)
top_movies = top_movies.sort_values(by='duration_min', ascending=False).head(5)

plt.figure(figsize=(8, 4))
sns.barplot(data=top_movies, x='duration_min', y='title')
plt.title("Top 5 des films les plus longs")
plt.xlabel("Dur√©e (minutes)")
plt.ylabel("Film")
plt.show()

# j. R√©partition des r√©alisateurs des ≈ìuvres fran√ßaises
fr_works = df[df['country'].str.contains("France", na=False)]
plt.figure(figsize=(10, 6))
fr_works['director'].dropna().value_counts().head(10).plot(kind='bar', color='darkred')
plt.title("Top 10 des r√©alisateurs d'≈ìuvres fran√ßaises")
plt.xlabel("R√©alisateur")
plt.ylabel("Nombre d'≈ìuvres")
plt.xticks(rotation=45)
plt.show()

# k. R√©partition des ≈ìuvres par date d'ajout au catalogue
date_counts = pd.to_datetime(df['date_added'], errors='coerce').dt.date.value_counts().sort_index()

plt.figure(figsize=(14, 5))
date_counts.plot()
plt.title("R√©partition des ajouts au catalogue au fil du temps")
plt.xlabel("Date d'ajout")
plt.ylabel("Nombre d'ajouts")
plt.tight_layout()
plt.show()

# l. Dur√©e moyenne des films selon le genre
films = df[df['type'] == 'Movie'].copy()
films['duration_min'] = films['duration'].str.extract(r'(\d+)').astype(float)
films['genre'] = films['listed_in'].str.split(', ').str[0]

plt.figure(figsize=(12, 5))
films.groupby('genre')['duration_min'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='teal')
plt.title("Dur√©e moyenne des films selon le genre")
plt.ylabel("Dur√©e moyenne (min)")
plt.xlabel("Genre")
plt.xticks(rotation=45)
plt.show()
```

### Analyse des graphiques de visualisation du dataset Netflix

**a. Type d'≈ìuvres :**
Le dataset contient principalement des films, mais les s√©ries repr√©sentent aussi une part significative. Cela montre que Netflix mise sur les deux formats, avec toutefois une pr√©dominance des films.

**b. Pays d'origine :**
Les ≈ìuvres proviennent majoritairement des √âtats-Unis, suivis de l'Inde, du Royaume-Uni et du Canada. La France arrive en cinqui√®me position. Cela d√©montre une forte dominance des pays anglophones, avec une certaine diversit√© internationale.

**c. Ann√©e d'ajout au catalogue :**
On observe une croissance importante des ajouts √† partir de 2016, avec un pic en 2020. Cela correspond √† la strat√©gie de croissance de Netflix, renforc√©e par la crise du COVID-19.

**d. Ratings :**
Les ≈ìuvres sont principalement destin√©es √† un public adulte (TV-MA) ou adolescent (TV-14). Les contenus pour enfants ou familles sont moins fr√©quents.

**e. Dur√©e des films :**
La plupart des films durent entre 80 et 120 minutes, ce qui correspond aux standards classiques du cin√©ma. Peu de films d√©passent les 2 heures.

**f. Dur√©e des s√©ries (saisons) :**
La majorit√© des s√©ries ont une ou deux saisons. Il y a peu de s√©ries tr√®s longues, ce qui sugg√®re une rotation rapide du contenu ou des formats courts.

**g. Genres :**
Les genres les plus repr√©sent√©s sont "Dramas", "Comedies" et "International Movies". Cela montre un int√©r√™t pour les contenus narratifs et diversifi√©s culturellement.

**h. Top 5 des s√©ries les plus longues :**
Certaines s√©ries atteignent jusqu'√† 15 saisons. Ces cas restent rares mais montrent que quelques titres phares restent longtemps dans le catalogue.

**i. Top 5 des films les plus longs :**
Des films de plus de 200 minutes existent, mais ils sont exceptionnels. Cela confirme que Netflix pr√©f√®re le format moyen pour le grand public.

**j. R√©alisateurs fran√ßais :**
Quelques r√©alisateurs se d√©marquent avec plusieurs ≈ìuvres disponibles. Cela peut r√©v√©ler des partenariats entre Netflix et certains auteurs locaux.

**k. Date d'ajout :**
Les ajouts sont r√©guliers et en augmentation jusqu'√† 2020. Ensuite, on observe un ralentissement. Cela peut √™tre li√© √† des changements strat√©giques ou √† l'effet post-pand√©mie.

**l. Dur√©e moyenne des films par genre :**
Certains genres comme le drame ou le documentaire pr√©sentent des dur√©es moyennes plus longues. Cela refl√®te des besoins narratifs plus complexes.

**Conclusion g√©n√©rale :**
Netflix dispose d'un catalogue diversifi√©, majoritairement compos√© de films et fortement orient√© vers les pays anglophones. Les ajouts ont √©t√© tr√®s dynamiques entre 2016 et 2020, avec une strat√©gie centr√©e sur des formats accessibles et populaires.
